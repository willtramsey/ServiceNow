### Unpacking the rest of the dictionaries

```python
# Ok, so now let's unpack the other dictionaries into new series, and add them to main df
# adding try/except blocks here so that the code is idempotent (because I 'run all' a lot)

try:
    if df_main['opened_by_value'] is None:
        pass
except:
    df_opened_by = df_main.opened_by.apply(pd.Series)
    df_opened_by['opened_by_value'] = df_opened_by['value']
    df_main = pd.concat([df_main, df_opened_by['opened_by_value']],axis=1)


try:
    if df_main['sys_domain_value'] is None:
        pass
except:    
    df_sys_domain = df_main.sys_domain.apply(pd.Series)
    df_sys_domain['sys_domain_value'] = df_sys_domain['value']
    df_main = pd.concat([df_main, df_sys_domain['sys_domain_value']],axis=1)


try:
    if df_main['configuration_item_value'] is None:
        pass
except:      
    df_cmdb_ci = df_main.cmdb_ci.apply(pd.Series)
    df_cmdb_ci['configuration_item_value'] = df_cmdb_ci['value']
    df_main = pd.concat([df_main, df_cmdb_ci['configuration_item_value']],axis=1)

    
try:
    if df_main['caller_id_value'] is None:
        pass
except:  
    df_caller_id = df_main.caller_id.apply(pd.Series)
    df_caller_id['caller_id_value'] = df_caller_id['value']
    df_main = pd.concat([df_main, df_caller_id['caller_id_value']],axis=1)

    
try:
    if df_main['assignment_group_value'] is None:
        pass
except:  
    df_assignment_group = df_main.assignment_group.apply(pd.Series)
    df_assignment_group['assignment_group_value'] = df_assignment_group['value']
    df_main = pd.concat([df_main, df_assignment_group['assignment_group_value']],axis=1)

    
try:
    if df_main['closed_by_value'] is None:
        pass
except:    
    df_closed_by = df_main.closed_by.apply(pd.Series)
    df_closed_by['closed_by_value'] = df_closed_by['value']
    df_main = pd.concat([df_main, df_closed_by['closed_by_value']],axis=1)

    
try:
    if df_main['parent_incident_value'] is None:
        pass
except: 
    df_parent_incident = df_main.parent_incident.apply(pd.Series)
    df_parent_incident['parent_incident_value'] = df_parent_incident['value']
    df_main = pd.concat([df_main, df_parent_incident['parent_incident_value']],axis=1)

    
try:
    if df_main['assigned_to_value'] is None:
        pass
except: 
    df_assigned_to = df_main.assigned_to.apply(pd.Series)
    df_assigned_to['assigned_to_value'] = df_assigned_to['value']
    df_main = pd.concat([df_main, df_assigned_to['assigned_to_value']],axis=1)

    
try:
    if df_main['location_value'] is None:
        pass
except: 
    df_location = df_main.location.apply(pd.Series)
    df_location['location_value'] = df_location['value']
    df_main = pd.concat([df_main, df_location['location_value']],axis=1)

    
print('Number of Rows: ', df_main.shape[0])
print('Number of Columns: ', df_main.shape[1])
df_main.head(1)
```
